```python
import requests, time
import pandas as pd
```

```python
query = """
SELECT DISTINCT ?title_url WHERE{
    ?title_url wdt:P31 wd:Q5398426;
                rdfs:label ?title.
                FILTER(LANG(?title)= 'en').
}
LIMIT 2000
"""
query_result =requests.get("https://query.wikidata.org/sparql", params={'query': query, 'format': 'json'}).json()
```

```python
title_url = []
for result in query_result['results']['bindings']:
    title_url.append(result['title_url']['value'].rsplit('/', 1)[1])
```

```python
data = []
for url in title_url:
    print(url)
    query = """
       SELECT ?title ?country_of_origin ?num_seasons ?start_time ?end_time (GROUP_CONCAT(DISTINCT ?cast; SEPARATOR=', ') AS ?cast_members) (GROUP_CONCAT(DISTINCT ?producer; SEPARATOR=', ') AS ?producers) 
        (GROUP_CONCAT(DISTINCT ?original_broadcaster; SEPARATOR=', ') AS ?original_broadcasters) (GROUP_CONCAT(DISTINCT ?distribution; SEPARATOR=', ') AS ?distributed_by)
        (GROUP_CONCAT(DISTINCT ?genre; SEPARATOR=', ') AS ?genres) (GROUP_CONCAT(DISTINCT ?language; SEPARATOR=', ') AS ?languages) 
        (GROUP_CONCAT(DISTINCT ?award_received; SEPARATOR=', ') AS ?awards_received) (GROUP_CONCAT(DISTINCT ?review_info; SEPARATOR=', ') AS ?review_scores)
        (GROUP_CONCAT(DISTINCT ?spin_off; SEPARATOR=', ') AS ?series_spin_off)
        WHERE {
          wd:""" + url + """ rdfs:label ?title.

          OPTIONAL{wd:""" + url + """ wdt:P495 [rdfs:label ?country_of_origin].
            FILTER(LANG(?country_of_origin)='en').}
          OPTIONAL{wd:""" + url + """ wdt:P2437 ?num_seasons.}
          OPTIONAL{wd:""" + url + """ wdt:P580 ?start_time.}
          OPTIONAL{wd:""" + url + """ wdt:P582 ?end_time.}
          OPTIONAL{wd:""" + url + """ wdt:P161 [rdfs:label ?cast].
            FILTER(LANG(?cast)='en').}
          OPTIONAL{wd:""" + url + """ wdt:P162 [rdfs:label ?producer].
            FILTER(LANG(?producer)='en').}
          OPTIONAL{wd:""" + url + """ wdt:P449 [rdfs:label ?original_broadcaster].
            FILTER(LANG(?original_broadcaster)='en').}
          OPTIONAL{wd:""" + url + """ wdt:P750 [rdfs:label ?distribution].
            FILTER(LANG(?distribution)='en').}
          OPTIONAL{wd:""" + url + """ wdt:P364 [rdfs:label ?language].
                   FILTER(LANG(?language)='en').}
          OPTIONAL{wd:""" + url + """ wdt:P166 [rdfs:label ?award_received].
                   FILTER(LANG(?award_received)='en').}
          OPTIONAL{wd:""" + url + """ wdt:P136 [rdfs:label ?genre].
                   FILTER(LANG(?genre)='en').}
          OPTIONAL{wd:""" + url + """ p:P444 [ps:P444 ?review_score;
                   pq:P447 [rdfs:label ?review_score_by]].
                   FILTER(LANG(?review_score_by)='en').
                   BIND(CONCAT(STR(?review_score), '-', STR(?review_score_by)) AS ?review_info).
                  }
          OPTIONAL{wd:""" + url + """ wdt:P2512 [rdfs:label ?spin_off].
                   FILTER(LANG(?spin_off)='en').}
          FILTER(LANG(?title)='en').
        }
        GROUP BY ?title ?country_of_origin ?num_seasons ?start_time ?end_time
    """
    query_result = requests.get("https://query.wikidata.org/sparql", params={'query': query, 'format': 'json'}).json()
    d = {}
    d[url] = query_result
    data.append(d)
    time.sleep(2)
```

```python
country_origin = []
num_episodes = []
num_seasons = []
title = []
selected_data = []
for d in data:
    x = {}
    k = list(d.keys())[0]
    x['id'] = k
    info = d[k]['results']['bindings']
    for i in info:
        if 'title' in i:
            x['title'] = i['title']['value']
        if 'country_of_origin' in i:
            x['country_of_origin'] = i['country_of_origin']['value']
        if 'num_seasons' in i:
            x['num_seasons'] = i['num_seasons']['value']
        if 'start_time' in i:
            x['start_time'] = i['start_time']['value']
        if 'end_time' in i:
            x['end_time'] = i['end_time']['value']
        if 'cast_members' in i:
            x['cast_members'] = i['cast_members']['value']
        if 'producers' in i:
            x['producers'] = i['producers']['value']
        if 'original_broadcasters' in i:
            x['original_broadcasters'] = i['original_broadcasters']['value']
        if 'distributed_by' in i:
            x['distributed_by'] = i['distributed_by']['value']
        if 'genres' in i:
            x['genres'] = i['genres']['value']
        if 'languages' in i:
            x['languages'] = i['languages']['value']
        if 'awards_received' in i:
            x['awards_received'] = i['awards_received']['value']
        if 'review_scores' in i:
            x['review_scores'] = i['review_scores']['value']
        if 'series_spin_off' in i:
            x['series_spin_off'] = i['series_spin_off']['value']
        
    selected_data.append(x)
```

```python
df = pd.DataFrame(selected_data)
df.set_index('id')
```

```python
df.to_json('wikidata.jsonl', indent=3, orient='records', force_ascii=False)
```
